```python
# Import libraries
import pandas as pd
#import matplotlib.pyplot as plt
import numpy as np
import random
# Crear lista de productos
products = ["video", "celulares", "audio", "grandes E.", "E.menores"]

# Create a list of dates within the last .....
import datetime
today = datetime.date.today()
dates = [today - datetime.timedelta(days=random.randint(0, 90)) for _ in range(10)]

# Create a list of random
sales_values = [random.randint(100, 1500) for _ in range(10)]
quantity = [random.randint(1, 4) for _ in range(4)]

df = pandas.DataFrame({
    "producto": random.choices(products, k=10),
    "cantidad": random.choices(quantity, k=10),
    "fecha": dates,
    "valor": sales_values,
})
# Print the DataFrame

print(df)
df
```

```python
df["valor_compras"] =  df["cantidad"] * df["valor"]
df
```
