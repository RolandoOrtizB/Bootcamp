```python
import pandas as pd
import numpy as np

# Crear un DataFrame de ejemplo con datos de proveedores
data = {
    'Calidad': [9, 8, 7, 6, 9, 8, 7, 7, 8, 9],
    'Costo': [5, 4, 5, 6, 4, 3, 4, 5, 3, 2],
    'Lead_Time': [7, 9, 7, 8, 6, 7, 8, 9, 5, 6]
}

df_proveedores = pd.DataFrame(data)

```

```python

# Calcular la matriz de correlación
matriz_correlacion = df_proveedores.corr()

print("Matriz de Correlación entre Proveedores:\n")
print(matriz_correlacion)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Visualizar la matriz de correlación
plt.figure(figsize=(8, 6)) # Establecer el tamaño del gráfico
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Matriz de Correlación entre Métricas de Proveedores")
plt.show()
```
